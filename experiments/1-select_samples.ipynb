# Select samples from iHMP IBD dataset

We are interested in samples that has both metagenomics and metatranscriptomics samples

```python
import pandas as pd
main_metadta = "../../data/IBD/hmp2_metadata.csv"

df = pd.read_csv(main_metadta)
df
```

```python
df['data_type'].unique()
```

```python
df_dna = df[df['data_type']=='metagenomics']
df_dna = df_dna[['External ID','Participant ID', 'interval_days', 'visit_num', 'diagnosis', 'sex', 'race', 'consent_age']]
df_dna
```

```python
df_rna = df[df['data_type']=='metatranscriptomics']
df_rna = df_rna[['External ID', 'Participant ID', 'interval_days', 'visit_num', 'diagnosis', 'sex', 'race', 'consent_age']]
df_rna
```

```python
df_merged = df_dna[df_dna['External ID'].isin(df_rna['External ID'])]
df_merged
```

```python
df_merged['race'].count()
```

```python
len(df_merged['Participant ID'].unique())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

os.makedirs('figures', exist_ok = True)

# Set a style for Seaborn
sns.set_style("whitegrid")

# Plot a histogram (distplot) using Seaborn
sns.histplot(y=df_merged['race'], kde=False)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of races')



plt.savefig('./figures/race_hist.png', bbox_inches='tight', dpi=900)
plt.show()
```

```python
# We see that races other than white are under-represented. Therefore, only white race will be selected in the analysis

df_merged = df_merged[df_merged['race']=='White'].reset_index(drop=True)
df_merged
```
